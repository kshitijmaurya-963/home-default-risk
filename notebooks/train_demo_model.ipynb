```python
import pandas as pd

# Load data
df = pd.read_csv("../data/processed/loan_risk_gold_full.csv")

# Basic info
print("="*50)
print("Dataset Shape")
print(df.shape)

print("\nColumns")
print(df.columns.tolist())

print("\nTarget Distribution")
print(df["target"].value_counts())

print("\nTarget Ratio")
print(df["target"].value_counts(normalize=True))

# Remove ID
X = df.drop(columns=["sk_id_curr", "target"])

# Target
y = df["target"]

print("\nFeature Shape")
print(X.shape)

print("\nTarget Shape")
print(y.shape)

# Detect column types
categorical_cols = X.select_dtypes(
    include=["object", "category"]
).columns.tolist()

numeric_cols = X.select_dtypes(
    exclude=["object", "category"]
).columns.tolist()

print("\nCategorical Columns:", len(categorical_cols))
print(categorical_cols)

print("\nNumeric Columns:", len(numeric_cols))
print(numeric_cols[:20])   # first 20

# Missing values
missing = X.isnull().sum()

missing = (
    missing[missing > 0]
    .sort_values(ascending=False)
)

print("\nTop Missing Columns")
print(missing.head(20))
```

```python
selected_features = [

    # Financial

    "amt_income_total",

    "amt_credit",

    "amt_annuity",

    "amt_goods_price",


    # Demographic

    "days_birth",

    "days_employed",

    "region_population_relative",


    # Bureau history

    "bureau_avg_credit",

    "bureau_recent_credit",

    "bureau_delinquency_count",

    "bureau_active_count",

    "bureau_record_count",


    # Previous applications

    "prev_app_count",

    "prev_avg_credit",

    "prev_avg_annuity",


    # Engineered

    "debt_to_income",

    "credit_utilization",

    "days_employed_ratio",

    "approval_rate"
]

print("Number of Features:", len(selected_features))

print(selected_features)
```

```python
hybrid_features = [

    # Financial

    "amt_income_total",

    "amt_credit",

    "amt_annuity",

    "amt_goods_price",


    # Demographic

    "days_birth",

    "days_employed",

    "days_registration",

    "days_id_publish",

    "days_last_phone_change",

    "region_population_relative",


    # External risk scores

    "ext_source_1",

    "ext_source_2",

    "ext_source_3",


    # Bureau

    "bureau_avg_credit",

    "bureau_recent_credit",

    "bureau_delinquency_count",

    "bureau_active_count",

    "bureau_record_count",


    # Previous applications

    "prev_app_count",

    "prev_avg_credit",

    "prev_avg_annuity",


    # Engineered

    "debt_to_income",

    "credit_utilization",

    "days_employed_ratio",

    "approval_rate"
]

print(len(hybrid_features))
```

```python
missing_features = [

    col

    for col in selected_features

    if col not in df.columns
]

print("Missing Features:")

print(missing_features)
```

```python
missing = [

    col

    for col in hybrid_features

    if col not in df.columns
]

print(missing)
```

```python
X_demo = df[selected_features]

y_demo = df["target"]

print(X_demo.shape)

print(y_demo.shape)
```

```python
X_hybrid = df[hybrid_features]

y_hybrid = df["target"]

print(X_hybrid.shape)
```

```python
missing_demo = (

    X_demo

    .isnull()

    .sum()

    .sort_values(ascending=False)
)

print(missing_demo)
```

```python
X_hybrid.isnull().sum().sort_values(
    ascending=False
)
```

```python
from sklearn.model_selection import train_test_split

X_train_demo, X_test_demo, y_train_demo, y_test_demo = train_test_split(
    X_demo,
    y_demo,
    test_size=0.2,
    stratify=y_demo,
    random_state=42
)

print(X_train_demo.shape)
print(X_test_demo.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train_hybrid, X_test_hybrid, y_train_hybrid, y_test_hybrid = train_test_split(

    X_hybrid,

    y_hybrid,

    test_size=0.2,

    stratify=y_hybrid,

    random_state=42

)

print(X_train_hybrid.shape)

print(X_test_hybrid.shape)
```

```python
n_negative = (y_train_demo == 0).sum()

n_positive = (y_train_demo == 1).sum()

scale_pos_weight_demo = n_negative / n_positive

print(scale_pos_weight_demo)
```

```python
n_negative = (y_train_hybrid == 0).sum()

n_positive = (y_train_hybrid == 1).sum()

scale_pos_weight_hybrid = n_negative / n_positive

print(scale_pos_weight_hybrid)
```

```python
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer

import lightgbm as lgb


demo_pipeline = Pipeline(

    steps=[

        (

            "imputer",

            SimpleImputer(

                strategy="median"

            )

        ),

        (

            "model",

            lgb.LGBMClassifier(

                objective="binary",

                metric="auc",

                n_estimators=500,

                learning_rate=0.05,

                num_leaves=63,

                scale_pos_weight=scale_pos_weight_demo,

                random_state=42,

                verbose=-1

            )

        )

    ]
)


print(demo_pipeline)
```

```python
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer

import lightgbm as lgb


hybrid_pipeline = Pipeline(

    steps=[

        (

            "imputer",

            SimpleImputer(

                strategy="median"

            )

        ),

        (

            "model",

            lgb.LGBMClassifier(

                objective="binary",

                metric="auc",

                n_estimators=500,

                learning_rate=0.05,

                num_leaves=63,

                scale_pos_weight=scale_pos_weight_hybrid,

                random_state=42,

                verbose=-1

            )

        )

    ]

)

print(hybrid_pipeline)
```

```python
from time import time

start = time()

demo_pipeline.fit(

    X_train_demo,

    y_train_demo

)

end = time()

print(

    f"Training Time: {end-start:.2f} sec"

)
```

```python
from time import time

start = time()

hybrid_pipeline.fit(

    X_train_hybrid,

    y_train_hybrid

)

end = time()

print(

    f"Training Time: {end-start:.2f} sec"

)
```

```python
from sklearn.metrics import (

    roc_auc_score,

    accuracy_score,

    precision_score,

    recall_score,

    f1_score,

    confusion_matrix

)


y_prob = demo_pipeline.predict_proba(

    X_test_demo

)[:,1]


y_pred = demo_pipeline.predict(

    X_test_demo

)


print("="*50)

print(

    "ROC AUC:",

    round(

        roc_auc_score(

            y_test_demo,

            y_prob

        ),

        4

    )

)


print(

    "Accuracy:",

    round(

        accuracy_score(

            y_test_demo,

            y_pred

        ),

        4

    )

)


print(

    "Precision:",

    round(

        precision_score(

            y_test_demo,

            y_pred

        ),

        4

    )

)


print(

    "Recall:",

    round(

        recall_score(

            y_test_demo,

            y_pred

        ),

        4

    )

)


print(

    "F1:",

    round(

        f1_score(

            y_test_demo,

            y_pred

        ),

        4

    )

)


print("\nConfusion Matrix")

print(

    confusion_matrix(

        y_test_demo,

        y_pred

    )

)
```

```python
from sklearn.metrics import (

    roc_auc_score,

    accuracy_score,

    precision_score,

    recall_score,

    f1_score,

    confusion_matrix

)

y_prob = hybrid_pipeline.predict_proba(

    X_test_hybrid

)[:,1]

y_pred = hybrid_pipeline.predict(

    X_test_hybrid

)


print("="*50)

print(

    "ROC AUC:",

    round(

        roc_auc_score(

            y_test_hybrid,

            y_prob

        ),

        4

    )

)

print(

    "Accuracy:",

    round(

        accuracy_score(

            y_test_hybrid,

            y_pred

        ),

        4

    )

)

print(

    "Precision:",

    round(

        precision_score(

            y_test_hybrid,

            y_pred

        ),

        4

    )

)

print(

    "Recall:",

    round(

        recall_score(

            y_test_hybrid,

            y_pred

        ),

        4

    )

)

print(

    "F1:",

    round(

        f1_score(

            y_test_hybrid,

            y_pred

        ),

        4

    )

)

print("\nConfusion Matrix")

print(

    confusion_matrix(

        y_test_hybrid,

        y_pred

    )

)
```
